```python
figure = plt.figure(figsize=(3*5, 5))
axes1 = figure.add_subplot(1,3,1)
axes1.plot(x, np.cos(x))
axes1.grid(lw=2)

axes2 = figure.add_subplot(1,3,2)
axes2.plot(x, np.sin(x))
axes2.grid(color='r')

axes3  = figure.add_subplot(1,3,3)
axes3.plot(x, x**2)
axes3.grid(c='g', ls='--')
# 子图没有title方法
# axes3.title('y=x^2')
axes3.set_title('y=x^2', )
```

### 保存图片

figure.savefig的选项
+ filename  
含有文件路径的字符串或Python的文件型对象。图像格式由文件扩展名推断得出，例如，.pdf推断出PDF，.png推断出PNG
（“png”、“pdf”、“svg”、“ps”、“eps”……）
+ dpi  
图像分辨率（每英寸点数），默认为100
+ facecolor  
图像的背景色，默认为“w”（白色） 

```python
figure = plt.figure(figsize=(10,6))
n1 = np.random.normal(loc=0, scale=2,size=100)
plt.plot(n1, label='one', linestyle='-.')

n2 = np.random.normal(loc=10, scale=4, size=100)
plt.plot(n2, label='two', ls='--')

n3 = np.random.normal(loc=20, scale=5, size=100)
plt.plot(n3, label='3rd', ls=':')

plt.legend()

figure.savefig('./art.png', facecolor='g', dpi=200, transparent=True, edgecolor='r')
```

### 直方图

【直方图的参数只有一个x！！！不像条形图需要传入x,y】

hist()的参数
+ bins  
可以是一个bin数量的整数值，也可以是表示bin的一个序列。默认值为10
+ density  
如果值为True，直方图的值将进行归一化处理，形成概率密度，默认值为False
+ color  
指定直方图的颜色。可以是单一颜色值或颜色的序列。如果指定了多个数据集合，颜色序列将会设置为相同的顺序。如果未指定，将会使用一个默认的线条颜色
+ orientation  
通过设置orientation为horizontal创建水平直方图。默认值为vertical

```python
x = np.random.randint(0,10, size=10)
x
```

```python
plt.hist(x, bins=50, density=True, color='r', orientation='horizontal')
```
