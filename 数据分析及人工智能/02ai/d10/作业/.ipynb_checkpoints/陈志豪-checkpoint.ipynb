# 城市气候与海洋的关系研究

导入包

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from pandas import Series,DataFrame
import glob
%matplotlib inline
```

导入数据各个海滨城市数据

```python
files = glob.glob(os.path.join('./', "*.csv"))
files
```

```python
dl= []
for f in files:
    dl.append(pd.read_csv(f))
df=pd.concat(dl)
```

```python
df
```

查看行数

```python
df.shape
```

去除没用的列

```python
df = df.drop(['dt','description','day','Unnamed: 0'],axis=1)
```

```python
df.head()
```

各城市与海洋距离，最高温度，最低温度，最高湿度，最低湿度

```python
df['city'].unique()
```

```python
max_temp = df.groupby('city')[['temp']].max()
max_temp = max_temp.rename(columns={'temp':'max_temp'})
min_temp = df.groupby('city')[['temp']].min()
min_temp = min_temp.rename(columns={'temp':'min_temp'})
max_hum = df.groupby('city')[['humidity']].max()
max_hum = max_hum.rename(columns={'humidity':'max_hum'})
min_hum = df.groupby('city')[['humidity']].min()
min_hum = min_hum.rename(columns={'humidity':'min_hum'})
dist = df.groupby('city')[['dist']].mean()
display(max_temp,min_temp,max_hum,min_hum,dist)
```

```python
df2 = pd.concat([dist,max_temp,min_temp,max_hum,min_hum],axis=1)
df2
```

显示最高温度与离海远近的关系

```python
df2.plot('dist','max_temp',kind='scatter')
```

可以看到散点图显示效果不好

由于风向是360度，我们可以考虑使用玫瑰图（极坐标条形图）

首先自定义一个画图函数

```python
def draw(x,y):
    plt.figure(figsize=(8,8))
    plt.subplot(projection='polar',facecolor='g')
    plt.bar(x,y,color=np.random.rand(8,3),width=0.2)
```

用numpy创建一个直方图，将360度划分为8个面元，将数据分类到这8个面元中

计算米兰各个方向上的风速

```python
df.head()
```

```python
con = df['city'] == 'Milano'
ml_wind_speed = df[con].groupby(by='wind_deg')['wind_speed'].mean()
ml_wind_speed
```

```python
deg = list(ml_wind_speed.index)
deg
```

将各个方向的风速保存在列表中

```python
wind_speed = list(ml_wind_speed.values)
wind_speed
```

画出各个方向的风速

```python
draw(deg,wind_speed)
```

```python
df.head()
```

将上面步骤写成函数

```python
def city_wind(city):
    con = df['city'] == city
    wind_mean_speed = df[con].groupby(by='wind_deg')['wind_speed'].mean()
    w_deg = list(wind_mean_speed.index)
    w_speed = list(wind_mean_speed.values)
    plt.figure(figsize=(8,8))
    plt.subplot(projection='polar',facecolor='g')
    plt.bar(w_deg,w_speed,color=np.random.rand(8,3),width=0.2)
```

```python
city_wind('Asti')
```

### 2、小麦种类预测
读取seeds.tsv文件，最后一列是小麦品种，其他列是小麦特征

```python
df_seed = pd.read_csv('./seeds.tsv',sep='\t',header=None)
df_seed.head()
```

```python
seed_cha = df_seed.values[:,:-1]
seed_var = df_seed.values[:,-1]
display(seed_cha,seed_var)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train,X_test,y_train,y_test = train_test_split(seed_cha,seed_var,test_size=0.2)
```

```python
X_train,X_test,y_train,y_test
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
y_ = knn.predict(X_test)
y_
```

```python
knn.score(X_test,y_test)
```

```python
knn.score(X_train,y_train)
```

### 3、改进约会网站的匹配效果
读取datingTestSet.txt文件，最后一列是喜欢程度。模型：根据前几列的信息，预测喜欢程度

```python
data_df = pd.read_table('./datingTestSet.txt',header=None)
data_df.head()
```

```python
data = data_df.values[:,:-1]
like = data_df.values[:,-1]
display(data,like)
```

```python
X_train,X_test,y_train,y_test = train_test_split(data,like,test_size=0.2)
X_train,X_test,y_train,y_test
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
y_ = knn.predict(X_test)
```

```python
y_
```

```python
knn.score(X_test,y_test)
```

```python
knn.score(X_train,y_train)
```
